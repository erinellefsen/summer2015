```python
import graph
import random
import pandas as pd
import matplotlib.pyplot as plt
import pylab as pl
import numpy as np
import collections
%matplotlib inline
```

```python
#NuM infected, not proportion
vaccination = 0
colorlist = ['gray','lightblue','red','lightgreen']
for color in colorlist:
    population = 500
    probofconnlist = [.01,.005,.00333,.0025,.002]
    for y in probofconnlist:
        RandIdict = []

        k = 5

        for x in range(10):
            g = graph.Graph(k,.7,0,vaccination*population)
            g.makeVertices(population)
            g.makeConnections(y)
            g.update()
            r = g.calculateR(True)
            i = g.getR()
            RandIdict = RandIdict + [[r, float(i),population]]
            k += 1

        p = .1
        for x in range(9):
            g = graph.Graph(8,p,0,vaccination*population)
            g.makeVertices(population)
            g.makeConnections(y)
            g.update()
            r = g.calculateR(True)
            i = g.getR()
            RandIdict = RandIdict + [[r, float(i),population]]
            p += .1
        
        population += 500


    
    print(RandIdict)
    Rlist = []
    Ilist = []

    for item in RandIdict:
        Rlist = Rlist + [item[0]]
        Ilist = Ilist + [item[1]]
        #print(item,RandIdict[item])
    
    plt.scatter(Rlist,Ilist,s=50,c=color)

    vaccination += .2



```
